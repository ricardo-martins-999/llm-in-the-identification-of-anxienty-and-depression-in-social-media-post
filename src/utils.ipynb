```python
import pandas as pd
```

```python
def read_dataset(path):
  return pd.read_excel(path)

def filter_dataset(dataset: pd.DataFrame, label: str, qtde_register: int = 50):
  dataset_filter = dataset[dataset['rotulo_humano'] == label]
  return dataset_filter.sample(n=qtde_register)

def merge_datasets(dataset_symptom, dataset_no_symptom):
  return pd.concat([dataset_symptom, dataset_no_symptom])

def export_dataset(dataset):
  dataset.to_excel('drive/MyDrive/Notebooks/Others/Sub Dataset/sub-dataset.xlsx', index=False)
```

```python
dataset = read_dataset('drive/MyDrive/Notebooks/Bert/dataset/dataset-pt.xlsx')
dataset_symptom = filter_dataset(dataset, 'sintoma')
dataset_no_symptom = filter_dataset(dataset, 'sem_sintoma')
dataset = merge_datasets(dataset_symptom, dataset_no_symptom)
export_dataset(dataset)
```
